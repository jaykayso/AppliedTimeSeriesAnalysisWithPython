# United States Macroeconomic Dataset

For this project we are going to predict realgdp and everything else is going to be considered exogenous variables / features to help us predict. 


```python
macro_data = sm.datasets.macrodata.load_pandas()
```

Storing the data.

```python
macro_data = macro_data.data
```

### Visualizing the Data

We are not going to be plotting year and quarter. axes.flatten is going to multiply the rows by columns so that we will be enumerating 12 times.
We are adding 2 to i in order to skip the first 2 columns of time periods that we don't want.

Looking at these graphs helps us to understand what each variable is doing over time. 

```python
fig, axes = plt.subplots(nrows=6, ncols=2, dpi=240, figsize = (10,10))

for i, ax in enumerate(axes.flatten()):
    data = macro_data[macro_data.columns[i+2]]
    
    ax.plot(data, color='black', linewidth = 1)
    ax.set_title(macro_data.columns[i+2])
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    ax.spines['top'].set_alpha(0)
    ax.tick_params(labelsize=6)
    
plt.tight_layout()
```

### Making the Data Stationary with Differencing

```python
macro_data['realgdp_tr1'] = macro_data['realgdp'].diff()

macro_data.head(10)
```

### Testing for Stationarity

```python
plt.plot(macro_data['realgdp_tr1'])
plt.title('Transformed Real GDP')
plt.show();
```

### AD Fuller Test for Stationarity

Yes, our new series is stationary. 

```python
ad_fuller_result = adfuller(macro_data['realgdp_tr1'][1:])

print(f'ADF Statistic: {ad_fuller_result[0]}')
print(f'p-value: {ad_fuller_result[1]}')
```

### Plotting ACF & PACF To Determine Order on AR or MA Process

They both result in a sinusoidal shape so there are no indicators of order that we can derive from these. 

```python
plot_acf(macro_data['realgdp_tr1'][1:]);
plot_pacf(macro_data['realgdp_tr1'][1:]);
```

### Running the Stationary Data Through an Optimization Function to Determine Best Model Coefficients 

```python
def optimize_SARIMAX(endog, exog, parameters_list, d, D, s):
    '''
        Returns dataframe with parameters, corresponding AIC
        
        endog - the observed variable
        exog - the exogenous variables
        parameters_list - list with (p,d,P,Q) tuples
        d - integration order
        D - seasonal integration order
        s - length of the season
    '''
    
    results = []
    
    for param in tqdm_notebook(parameters_list):
        try: 
            model = SARIMAX(endog,
                           exog, 
                           order = (param[0], d, param[1]),
                           seasonal_order=(param[2], D, param[3], s),
                           simple_differencing = False).fit(disp=False)
        except: 
            continue
        
        aic = model.aic
        results.append([param, aic])
    
    result_df = pd.DataFrame(results)
    result_df.columns = ['(p,q)x(P,Q)', 'AIC']
    result_df = result_df.sort_values(by='AIC', ascending = True).reset_index(drop = True)
    
    return result_df
        
```

Defining List of Params

```python
p = range(0,4,1) #0,1,2,3
d = 1
q = range(0,4,1)
P = range(0,4,1)
D = 0 #no seasonal diff
Q = range(0,4,1)
s = 4

parameters = product( p, q, P, Q)
parameters_list = list(parameters)

parameters_list
```

Optimize the function. We are only going to row 200 in order to separate the data for testing. 

```python
endog = macro_data['realgdp'][:200] #separation for training & testing
exog = macro_data[['realcons', 'realinv', 'realgovt', 'realdpi', 'cpi', 'm1', 'tbilrate', 'unemp', 'pop', 'infl', 'realint']][:200]

result_df = optimize_SARIMAX(endog, exog, parameters_list, 1, 0, 4)

result_df
```

Now we can refit the best model with:

- p = 2
- q = 2
- P = 0
- Q = 0

```python
best_model = SARIMAX(endog,
                     exog,
                     order = (2,0,2),
                    seasonal_order = (0,0,0,4), 
                    simple_differencing = False)

res = best_model.fit(dis=False)

print(res.summary())
```

### Residual Analysis

Not perfect, but fairly normal. 

```python
res.plot_diagnostics();
```

### Making Predictions 

Here the exogenous variable calling on -4 is predicting the data that we didn't use to train the model. We only went to row 200 for training data. We know that our forecasts are fairly accurate as the plot that we initially ran showed the dip at the end of the dataset. 

```python
n_forecast = 3 #.75 of year into future. This is the data we left out of the dataset.
predict = res.get_prediction(end=best_model.nobs + n_forecast, 
                            exog = exog.iloc[-4:])

idx = np.arange(len(predict.predicted_mean))
predict_ci = predict.conf_int(alpha = .5)

fig, ax = plt.subplots()
ax.plot(macro_data['realgdp'][:199], color='blue')
ax.plot(idx[-n_forecast:], predict.predicted_mean[-n_forecast:], 'k--')
ax.fill_between(idx, predict_ci.iloc[:,0], predict_ci.iloc[:,1], alpha=0.15)

ax.set_title('Forecast of US Real GDP')

plt.show();




```

### Where is the confidence interval? It is very small.

The pale gray line is the confidence interval. We had to zoom in. 

```python
fig, ax = plt.subplots()
ax.plot(macro_data['realgdp'], color = 'blue')
ax.plot(idx[-n_forecast:], predict.predicted_mean[-n_forecast:], 'k--')
ax.plot(predict.predicted_mean, 'ro')
ax.fill_between(idx, predict_ci.iloc[:,0], predict_ci.iloc[:,1], alpha = .15)
ax.set_xlim(190,199)
ax.set_ylim(13000,14100)

ax.set_title('Forecast of RealGDP')
plt.show();
```

### Calculating MSE

Lower than the one in the class. 

```python
mse = mean_squared_error(macro_data['realgdp'], predict.predicted_mean[:203])

print(f'MSE: {mse}')
```

### Calculating MAPE

It is difficult to interpret the MSE sometimes as it is a large number that can be somewhat confusing to conceptualize, so instead we can use Mean Percentage Error which will give us a percentage rather than an absolute number. 

Higher than the one in the class (39%)

```python
mape = np.mean(np.abs(macro_data['realgdp'] - predict.predicted_mean[:203]) / np.abs(macro_data['realgdp']))
```

```python
print(f'MAPE: {round(mape * 100 , 2)}%')
```
