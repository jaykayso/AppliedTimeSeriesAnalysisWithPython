# Autoregressive Model

### Importing Relevant Packages

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.regression.linear_model import yule_walker
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (10, 7.5)
```

### Simulating the Autoregressive process of order 2 

$ y_t = 0.33y_{t-1} + 0.5y_{t-2} $

First, we will initialize our array of coefficients.

```python
ar2 = np.array([1, 0.33, 0.5])

# we are putting zeros in the array to get rid of the moving average portion
ma2 = np.array([1, 0, 0])
```

```python
AR2_process = ArmaProcess(ar2, ma2).generate_sample(nsample=1000)
```

```python
plt.plot(AR2_process)
plt.title('Simulated AR(2) Process');
plt.xlim(0,200);
plt.show();
```

Now, we are going to look at the ACF. 

```python
plot_acf(AR2_process);
```

The oscillation is indicative of AR2 and not providing any info. Next we will do PACF.

```python
plot_pacf(AR2_process);
```

We can wee that the order here is 2 because the last significant peak is at point 2. Now we will try to model back to find the original coeficients. 

This time we will use the Yule Walker process.

```python
rho, sigma = yule_walker(AR2_process, 2, method='mle')
print(f'rho: {-rho}')
print(f'sigma:{sigma}')
```

The rho coeficients are close to .33 and .5. The sigma is close to 1. 

### Simulate AR(3) Process

Now we will do another simulation of AR(3)

$ y_t = 0.33y_{t-1} + 0.5y_{t-2} + .07y_{t-3} $

First, we need to initialize our coefficients. 

```python
ar3 = np.array([1, 0.33, 0.5, 0.07])

# we are putting zeros in the array to get rid of the moving average portion
ma3 = np.array([1, 0, 0, 0])
```

```python
AR3_process = ArmaProcess(ar3, ma3).generate_sample(nsample=1000)
```

```python
plt.plot(AR3_process)
plt.title('Simulated AR(3) Process');
plt.xlim(0,200);
plt.show();
```

### ACF

```python
plot_acf(AR3_process);
```

No new information can be pulled from this ACF plot. 

### PACF

```python
plot_pacf(AR3_process);
```

Here, we should be able to derive the order from the PACF graph however the barely significant point at x value 3 could be easily missed and could be misinterpreted as an order of 2. 

### Returning to the Equation

```python
rho, sigma = yule_walker(AR3_process, 3, method='mle')
print(f'rho: {-rho}')
print(f'sigma:{sigma}')
```

It did not do great on the coefficient of 0.07
