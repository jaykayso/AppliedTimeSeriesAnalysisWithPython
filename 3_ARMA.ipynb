# Autoregressive Moving Average (ARMA)

### Importing Relevant Packages

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_process import ArmaProcess

import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [10,7.5]
```

### Simulate ARMA(1,1)

```python
ar1 = np.array([1, 0.33])
ma1 = np.array([1, .9])
```

```python
ARMA_1 = ArmaProcess(ar1, ma1).generate_sample(nsample=1000)
```

```python
plt.plot(ARMA_1)
plt.title('Simulated ARMA(1,1) Process')
plt.xlim([0,200])
plt.show();
```

### Determining Order

When we have an ARMA process, we can't use ACF or PACF to determine the order. 

```python
plot_acf(ARMA_1);
```

```python
plot_pacf(ARMA_1);
```

This isn't consistent with the order of one that we input! 

### Simulate ARMA(2,2) Process

```python
ar2 = np.array([1, .33, .51])
ma2 = np.array([1, .9, .3])
```

```python
ARMA_2 = ArmaProcess(ar2, ma2).generate_sample(nsample=1000)
```

```python
plt.plot(ARMA_2)
plt.title('Simulated ARMA(2,2) Process')
plt.xlim([0,200])
plt.show();
```

```python
plot_acf(ARMA_2);
plot_pacf(ARMA_2);
```

This isn't consistent with the order of 2 that we input!
