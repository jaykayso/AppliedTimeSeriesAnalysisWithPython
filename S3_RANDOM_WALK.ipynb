### Importing Relevant Packages

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import acf
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

Setting the figure size for the notebook. 

```python
plt.rcParams['figure.figsize'] = [10 , 7.5]
```

### Simulation of Random Walk

We are going to assume that the random walk model starts at zero. Which means that we will use the formula from the notes that includes the sum of all of the errors. 

The following code will provide us with 1000 generated random error data points based off of the previous step.

```python
steps = np.random.standard_normal(1000)
steps[0] = 0

random_walk = np.cumsum(steps)
```

Looking at 10 data points. 

```python
random_walk[:10]
```

Plotting the random walk model. Since this is random, it will vary from person to person and run to run as well. 

```python
plt.plot(random_walk)
plt.title('Simulated Random Walk')
plt.show()
```

Now, let's take a look at the ACF function. We can actually extract the coefficients of the autocorrelation function. 

```python
random_walk_acf_coef = acf(random_walk)
random_walk_acf_coef
```

```python
plot_acf(random_walk, lags = 20);
```

Clearly we have autocorrelation in the plot above. So, our simulated random walk, is not a stationary time series. How can we make it stationary? We are currently taking the cumulative sum at each step. If we take the diference at each step, it should result in a random error. 

Finding the Random Error

```python
# n = 1 here means for every data point
random_walk_diff = np.diff(random_walk, n = 1)
```

Now we should have a stationary differenced time series. 

```python
plt.plot(random_walk_diff)
plt.title('Noise')
plt.show()
```

Now, let's take a look at the ACF plot for random noise. The ACF plot is indicating that the plot is now stationary as non of the points are outside of the area of significance.

```python
plot_acf(random_walk_diff, lags = 20);
```
